```python
#plot daily engagement
# Define engagement categories (e.g., users active 0-5 days, 6-10 days, etc.)
def categorize_days(days_active):
    if days_active == 0:
        return '0 Days'
    elif days_active <= 5:
        return '1-5 Days'
    elif days_active <= 10:
        return '6-10 Days'
    elif days_active <= 15:
        return '11-15 Days'
    else:
        return '16+ Days'

# Apply the categorization
activity_categories = activity_days_active.apply(categorize_days)
sleep_categories = sleep_days_active.apply(categorize_days)
weight_categories = weight_days_active.apply(categorize_days)

# Count users in each category
activity_counts = activity_categories.value_counts()
sleep_counts = sleep_categories.value_counts()
weight_counts = weight_categories.value_counts()
```

```python
import matplotlib.pyplot as plt

# Function to plot a donut chart with legend and percentages
def plot_donut_chart(ax, data, title):
    wedges, texts, autotexts = ax.pie(data,
                                       autopct='%1.1f%%',  # Format for percentage display
                                       startangle=90,
                                       wedgeprops=dict(width=0.4))  # Set width for the donut

    plt.setp(autotexts, size=16, weight="bold", color="black")  # Customize text appearance

    # Draw the center circle
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')  # Change the radius to adjust the hole size
    ax.add_artist(centre_circle)

    ax.set_title(title)
    ax.axis('equal')  # Equal aspect ratio ensures the donut chart is circular.

    # Create legend
    ax.legend(wedges, data.index, title="Engagement Categories", loc="center left", bbox_to_anchor=(0,0.5 , 0, 1))

# Create subplots for horizontal layout
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Plot donut charts in each subplot
plot_donut_chart(axs[0], activity_counts, 'User Engagement: Activity')
plot_donut_chart(axs[1], sleep_counts, 'User Engagement: Sleep')
plot_donut_chart(axs[2], weight_counts, 'User Engagement: Weight')

# Adjust layout
plt.tight_layout()
plt.show()


```

```python
#ACTIVITY METRICS AND Engagement metrics
import pandas as pd
import numpy as np

def analyze_user_engagement(activity_df, sleep_df, weight_df):
    """
    Analyzes user engagement patterns across different health recommendation categories

    Parameters:
    activity_df: DataFrame with daily activity data
    sleep_df: DataFrame with sleep records
    weight_df: DataFrame with weight measurements

    Returns:
    dict: Analysis results by health recommendation categories
    """
    activity_df['ActivityDate'] = pd.to_datetime(activity_df['ActivityDate'])
    sleep_df['Date'] = pd.to_datetime(sleep_df['Date'])
    if not weight_df.empty:
        weight_df['Date_date'] = pd.to_datetime(weight_df['Date_date'])
    analysis = {}

    # Calculate total possible days from the data range
    total_possible_days = {
        'activity': (activity_df['ActivityDate'].max() - activity_df['ActivityDate'].min()).days + 1,
        'sleep': (sleep_df['Date'].max() - sleep_df['Date'].min()).days + 1,
        'weight': (weight_df['Date_date'].max() - weight_df['Date_date'].min()).days + 1 if not weight_df.empty else 0
    }

    # 1. Activity Data Analysis
    analysis['activity'] = {}

    user_activity = activity_df.groupby('Id').agg({
        'TotalSteps': 'mean',
        'VeryActiveMinutes': 'mean',
        'FairlyActiveMinutes': 'mean',
        'LightlyActiveMinutes': 'mean',
        'SedentaryMinutes': 'mean',
        'Calories': 'mean'
    }).round(2)

    # Calculate logging percentage instead of days
    user_activity_days = activity_df.groupby('Id').size()
    user_activity['LoggingPercentage'] = (user_activity_days / total_possible_days['activity'] * 100).round(2)

    user_activity['StepsCategory'] = user_activity['TotalSteps'].apply(
        lambda x: 'Meeting Steps Rec.' if x >= 10000 else 'Not Meeting Steps Rec.'
    )

    activity_segments = user_activity.groupby('StepsCategory').agg({
        'TotalSteps': ['mean', 'min', 'max', 'count'],
        'VeryActiveMinutes': 'mean',
        'FairlyActiveMinutes': 'mean',
        'LightlyActiveMinutes': 'mean',
        'SedentaryMinutes': 'mean',
        'Calories': 'mean',
        'LoggingPercentage': ['mean', 'min', 'max']  # Changed from DaysLogged
    }).round(2)

    analysis['activity']['segments'] = activity_segments.to_dict()
    analysis['activity']['user_distribution'] = (
        user_activity['StepsCategory'].value_counts(normalize=True) * 100
    ).to_dict()

    # 2. Sleep Data Analysis
    analysis['sleep'] = {}

    user_sleep = sleep_df.groupby('Id').agg({
        'TotalMinutesAsleep': 'mean'
    }).round(2)

    user_sleep_days = sleep_df.groupby('Id').size()
    user_sleep['LoggingPercentage'] = (user_sleep_days / total_possible_days['sleep'] * 100).round(2)

    user_sleep['HoursAsleep'] = user_sleep['TotalMinutesAsleep'] / 60
    user_sleep['SleepCategory'] = user_sleep['HoursAsleep'].apply(
        lambda x: 'Meeting Sleep Rec.' if 7 <= x <= 9 else 'Not Meeting Sleep Rec.'
    )

    sleep_segments = user_sleep.groupby('SleepCategory').agg({
        'TotalMinutesAsleep': ['mean', 'min', 'max', 'count'],
        'HoursAsleep': ['mean', 'std'],
        'LoggingPercentage': ['mean', 'min', 'max']  # Changed from DaysLogged
    }).round(2)

    analysis['sleep']['segments'] = sleep_segments.to_dict()
    analysis['sleep']['user_distribution'] = (
        user_sleep['SleepCategory'].value_counts(normalize=True) * 100
    ).to_dict()

    # 3. Weight Data Analysis
    analysis['weight'] = {}

    if not weight_df.empty:
        weight_df['BMICategory'] = weight_df['BMI'].apply(
            lambda x: 'Normal BMI' if 18.5 <= x < 25 else 'Abnormal BMI'
        )

        user_weight = weight_df.groupby('Id').agg({
            'BMI': 'mean',
            'BMICategory': 'first',
            'WeightKg': 'mean'
        })

        user_weight_days = weight_df.groupby('Id').size()
        user_weight['LoggingPercentage'] = (user_weight_days / total_possible_days['weight'] * 100).round(2)

        weight_segments = user_weight.groupby('BMICategory').agg({
            'BMI': ['mean', 'min', 'max', 'count'],
            'WeightKg': 'mean',
            'LoggingPercentage': ['mean', 'min', 'max']  # Changed from DaysLogged
        }).round(2)

        analysis['weight']['segments'] = weight_segments.to_dict()
        analysis['weight']['user_distribution'] = (
            user_weight['BMICategory'].value_counts(normalize=True) * 100
        ).to_dict()

    # 4. Engagement comparison across categories
    activity_engagement = user_activity.groupby('StepsCategory').agg({
        'TotalSteps': ['count', 'mean'],
        'LoggingPercentage': ['mean', 'min', 'max']
    })

    analysis['engagement_comparison'] = {
        'steps': {
            'users_per_category': activity_engagement[('TotalSteps', 'count')].to_dict(),
            'avg_steps_per_category': activity_engagement[('TotalSteps', 'mean')].to_dict(),
            'logging_percentage_per_category': activity_engagement[('LoggingPercentage', 'mean')].to_dict(),
            'min_logging_percentage': activity_engagement[('LoggingPercentage', 'min')].to_dict(),
            'max_logging_percentage': activity_engagement[('LoggingPercentage', 'max')].to_dict()
        },
        'sleep': {
            'users_per_category': sleep_segments[('TotalMinutesAsleep', 'count')].to_dict(),
            'avg_hours_per_category': sleep_segments[('HoursAsleep', 'mean')].to_dict(),
            'logging_percentage_per_category': sleep_segments[('LoggingPercentage', 'mean')].to_dict(),
            'min_logging_percentage': sleep_segments[('LoggingPercentage', 'min')].to_dict(),
            'max_logging_percentage': sleep_segments[('LoggingPercentage', 'max')].to_dict()
        }
    }

    if not weight_df.empty:
        analysis['engagement_comparison']['weight'] = {
            'users_per_category': weight_segments[('BMI', 'count')].to_dict(),
            'avg_bmi_per_category': weight_segments[('BMI', 'mean')].to_dict(),
            'logging_percentage_per_category': weight_segments[('LoggingPercentage', 'mean')].to_dict(),
            'min_logging_percentage': weight_segments[('LoggingPercentage', 'min')].to_dict(),
            'max_logging_percentage': weight_segments[('LoggingPercentage', 'max')].to_dict()
        }

    return analysis

def print_engagement_report(analysis):
    """Prints a formatted report of the engagement analysis"""
    print("=== HEALTH METRICS AND ENGAGEMENT ANALYSIS ===\n")

    print("1. STEPS RECOMMENDATION ANALYSIS")
    print("User Distribution:")
    for category, pct in analysis['activity']['user_distribution'].items():
        print(f"- {category}: {pct:.1f}%")
    print(f"\nEngagement Metrics:")
    eng_stats = analysis['engagement_comparison']['steps']
    for category in eng_stats['users_per_category'].keys():
        print(f"\n{category}:")
        print(f"- Number of users: {eng_stats['users_per_category'][category]}")
        print(f"- Average steps: {eng_stats['avg_steps_per_category'][category]:,.0f}")
        print(f"- Logging percentage: {eng_stats['logging_percentage_per_category'][category]:.1f}%")
        print(f"  (Range: {eng_stats['min_logging_percentage'][category]:.1f}% - {eng_stats['max_logging_percentage'][category]:.1f}%)")

    print("\n2. SLEEP RECOMMENDATION ANALYSIS")
    print("User Distribution:")
    for category, pct in analysis['sleep']['user_distribution'].items():
        print(f"- {category}: {pct:.1f}%")
    print(f"\nEngagement Metrics:")
    sleep_stats = analysis['engagement_comparison']['sleep']
    for category in sleep_stats['users_per_category'].keys():
        print(f"\n{category}:")
        print(f"- Number of users: {sleep_stats['users_per_category'][category]}")
        print(f"- Average sleep: {sleep_stats['avg_hours_per_category'][category]:.1f} hours")
        print(f"- Logging percentage: {sleep_stats['logging_percentage_per_category'][category]:.1f}%")
        print(f"  (Range: {sleep_stats['min_logging_percentage'][category]:.1f}% - {sleep_stats['max_logging_percentage'][category]:.1f}%)")

    if 'weight' in analysis['engagement_comparison']:
        print("\n3. BMI CATEGORY ANALYSIS")
        print("User Distribution:")
        for category, pct in analysis['weight']['user_distribution'].items():
            print(f"- {category}: {pct:.1f}%")
        print(f"\nEngagement Metrics:")
        weight_stats = analysis['engagement_comparison']['weight']
        for category in weight_stats['users_per_category'].keys():
            print(f"\n{category}:")
            print(f"- Number of users: {weight_stats['users_per_category'][category]}")
            print(f"- Average BMI: {weight_stats['avg_bmi_per_category'][category]:.1f}")
            print(f"- Logging percentage: {weight_stats['logging_percentage_per_category'][category]:.1f}%")
            print(f"  (Range: {weight_stats['min_logging_percentage'][category]:.1f}% - {weight_stats['max_logging_percentage'][category]:.1f}%)")
```

```python
engagement_analysis = analyze_user_engagement(df_activity, df_sleep, df_weight)
print_engagement_report(engagement_analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import pandas as pd
# Import NullLocator from the correct module
from matplotlib.ticker import NullLocator


def plot_activity_engagement_stacked_area(activity_df):
    # Convert date column to datetime
    activity_df['ActivityDate'] = pd.to_datetime(activity_df['ActivityDate'])

    # Categorize users based on step recommendation
    activity_df['StepsCategory'] = activity_df['TotalSteps'].apply(lambda x: 'Meeting Steps Rec.' if x >= 10000 else 'Not Meeting Steps Rec.')

    # Calculate daily user logs per category
    daily_activity_logs = activity_df.groupby(['ActivityDate', 'StepsCategory'])['Id'].nunique().reset_index(name='UserCount')

    # Pivot data to make it suitable for stacked area plot
    daily_activity_logs_pivot = daily_activity_logs.pivot(index='ActivityDate', columns='StepsCategory', values='UserCount').fillna(0)

    # Plot stacked area chart
    plt.figure(figsize=(10, 10))
    daily_activity_logs_pivot.plot(kind='area', stacked=True, alpha=0.7, colormap='Paired')

    # Customize plot
    plt.title('User Engagement in Activity Tracking (Meeting vs Not Meeting Recommendations)')
    plt.ylabel('Number of Users')
    plt.xlabel('Date')

    # Format x-axis to show dates in a clear format
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.xticks(rotation=45)

    # Remove minor ticks on the x-axis to prevent overlap
    # Use NullLocator imported from matplotlib.ticker
    plt.gca().xaxis.set_minor_locator(NullLocator())

    plt.legend(title='Steps Category')
    plt.tight_layout()
    plt.show()

# Usage
plot_activity_engagement_stacked_area(df_activity)
```
